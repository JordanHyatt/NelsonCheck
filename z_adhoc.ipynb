```python
import random
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
DF = pd.DataFrame

from NelsonCheck import NelsonCheck as NC
```

```python
data = [random.random() for i in range(0,500)]


data = data + [.9,.9,.8,.9,.8,.9,.8,.9,.8,.9,.8,.9,.8,.9,.8,.7]
#data = pd.read_csv('bug.csv').data
self = NC.NelsonCheck(data=data, exclude_rules=[1,2,3,5,6,7,8])


df = DF()
data = self.data
df['ser'] = data
df['position'] = [i for i in range(len(df))]
df['prev'] = data.shift(1)
df['dif'] = df['ser'] - df['prev']
def get_label(val):
    if val>0:
        return 'up'
    elif val==0:
        return 'same'
    else:
        return 'down'
df['label'] = df.dif.apply(get_label)
df['nlabel'] = df['label'].shift(-1)
def get_same_dir(ser):
    m1 = ser.label != ser.nlabel
    m2 = ser.label != 'same'
    m3 = ser.nlabel != 'same'
    if m1 and m2 and m3:
        return False
    else:
        return True
df['same_dir']=df.apply(get_same_dir,axis=1)
df['rsum'] = df['same_dir'].rolling(12).sum()

df.iloc[-20:]
```
